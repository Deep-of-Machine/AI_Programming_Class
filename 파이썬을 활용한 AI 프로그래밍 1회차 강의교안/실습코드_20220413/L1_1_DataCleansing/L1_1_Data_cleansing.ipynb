# 결측치

```python
import numpy as np
from numpy import nan as NA
import pandas as pd

#np.random.seed(111)

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

# 일부 데이터를 일부러 누락시킵니다
sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA 

sample_data_frame
```

### 리스트와이즈

```python
sample_data_frame.dropna()
```

### 페어와이즈

```python
 sample_data_frame[[0,1,2]].dropna()
```

### 결측치 보완

```python
import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

# 일부 데이터를 일부러 누락시킵니다
sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA
```

### 0으로 채우기

```python
sample_data_frame.fillna(0)
```

### 앞에있는 데이터로 채우기

```python
sample_data_frame.fillna(method="ffill")
```

```python
sample_data_frame.fillna(method="bfill")
```

### 평균값 대입

```python
import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

# 일부 데이터를 일부러 누락시킵니다
sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[2, 2] = NA
sample_data_frame.iloc[5:, 3] = NA

#fillna로 NaN 부분에 열의 평균값을 대입합니다
sample_data_frame.fillna(sample_data_frame.mean())
```
